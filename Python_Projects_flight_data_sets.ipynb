```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```python
df= pd.read_csv("/content/flights_data.csv")
print(df.shape)    # shape how the dimension
df.head()       # it show the top 5 item in the data set.
```

```python
sb.countplot(data = df, x = 'Source');
```

```python
base_color = sb.color_palette()[3]
sb.countplot(data=df, x = 'Source', color = base_color);

```

```python
base_color= sb.color_palette()[2]
sorted_order = df['Source'].value_counts()
print(sorted_order)
```

```python
base_color= sb.color_palette()[2]
sorted_order = df['Source'].value_counts().index
sb.countplot(data = df, x ='Source', color = base_color, order = sorted_order);
```

```python
base_color = sb.color_palette()[0]
sb.countplot(data = df.head(), x = 'Route', color = base_color);
```

```python
base_color = sb.color_palette()[1]
sb.countplot(data = df.head(), x = 'Route', color= base_color);
plt.xticks(rotation = 90);
```

```python
base_color = sb.color_palette()[4]
sb.countplot(data = df.head(), y = 'Route', color= base_color);
#plt.xticks(rotation = 90);
```

```python
df.isna().sum()
```

```python
na_counts = df.isna().sum()
base_color = sb.color_palette()[2]
sb.barplot(na_counts.index.values,na_counts, color = base_color);
plt.xticks(rotation = 45)
```

```python
#pie_chart

sorted_counts = df['Source'].value_counts()
print(sorted_counts)
plt.pie(sorted_counts, labels = sorted_counts.index, startangle= 90, counterclock= False);
plt.axis('square');
```

```python
sorted_counts = df['Source'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle= 90, counterclock= False);
#plt.axis('square');
```

```python
# ringe pie- (donut)

sorted_counts = df['Source'].value_counts()
plt.pie(sorted_counts, labels = sorted_counts.index, startangle= 90, counterclock= False , wedgeprops = {'width': 0.4});
#plt.axis('square');
```
